**2. Explain the difference between mutable and immutable data types with examples ?**

## 1. Mutable Data Types
Mutable data types are those whose values **can be changed after the object is created**.  
The modification happens in the **same memory location**, so the original object is updated.
**Examples of Mutable Data Types**
- List
- Dictionary
- Set
## 2. Immutable Data Types
Immutable data types are those whose values **cannot be changed after the object is created**.  
If we try to change the value, **Python creates a new object instead of modifying the original one**.
**Examples of immutable Data structure**
- String
- Tuple

| Feature | Mutable | Immutable |
|--------|--------|--------|
| Meaning | Value can be changed | Value cannot be changed |
| Modification | Same object is modified | New object is created |
| Memory | Same memory location | New memory location |
| Examples | List, Dictionary, Set |  String, Tuple |

## Python Data Types

| List (`list`) | Integer (`int`) |
| Dictionary (`dict`) | Float (`float`) |
| Set (`set`) | Boolean (`bool`) |
|  | String (`str`) |
|  | Tuple (`tuple`) |

**8. What is a hash table, and how does it relate to dictionaries in Python?**

A **hash table** is a data structure that stores data in **key–value pairs**.  
It uses a special function called a **hash function** to convert a key into an index (a position in memory) where the value is stored.

This allows data to be **stored and retrieved very quickly**.

---

#### How Hash Tables Work

1. A **key** is given to the hash function.
2. The hash function converts the key into a **hash value (index)**.
3. The value associated with that key is stored at that index in memory.
4. When we search for a value, Python again uses the **same hash function** to quickly find the location.

---

#### Relation Between Hash Table and Dictionary in Python

A **dictionary**  is implemented using a **hash table**.

This means:

- Dictionaries store data as **key : value pairs**.
- Keys are converted into **hash values**.
- The hash value determines where the data is stored in memory.
- This makes dictionary operations like **searching, inserting, and deleting very fast**.

19. Why are dictionaries considered unordered, and how does that affect data retrieval ?

A **dictionary** in Python stores data as **key–value pairs**. Traditionally, dictionaries are considered **unordered** because the elements are stored using a **hash table**, not by their position like lists.

When data is inserted into a dictionary, Python calculates a **hash value** for each key and stores it in a location based on that hash. Because of this, the data is organized for **fast lookup**, not for maintaining order.

---

### Why Dictionaries Are Considered Unordered

- Dictionaries store data using **hashing**.
- Keys are placed in memory based on their **hash value**, not their insertion position.
- Therefore, the elements do not rely on numerical indexes.

Example:

```python
student = {
    "name": "Rajneesh",
    "age": 21,
    "course": "Data Analytics"
}

A **dictionary** in Python stores data in the form of **key–value pairs**. Traditionally, dictionaries are considered **unordered collections** because their elements are not arranged by position or index like lists or tuples. Instead, they are organized using a **hash table**, which focuses on fast data retrieval rather than maintaining a sequence.

---

### Why Dictionaries Are Considered Unordered

1. **Hash-based storage**  
   Python uses a **hashing mechanism** to store dictionary keys. The key is converted into a hash value, which determines where the data is stored in memory.

2. **No indexing system**  
   Unlike lists (`list[0]`, `list[1]`), dictionaries do not use numeric indexes to organize elements.

3. **Access through keys**  
   Each value is retrieved using its **unique key**, not its position.

> Note: In **Python 3.7 and later**, dictionaries maintain **insertion order**, but they are still conceptually treated as unordered structures because they are accessed by keys rather than indexes.

---

### How This Affects Data Retrieval

| Effect | Explanation |
|------|-------------|
| Access by key | Values are retrieved using keys, e.g., `dict["name"]`. |
| No positional access | You cannot access items using indexes like `dict[0]`. |
| Faster lookup | Hash tables allow very fast lookup, usually **O(1) time complexity**. |
| Unique keys | Each key must be unique, ensuring accurate retrieval of values. |

---

### Example

```python
student = {
    "name": "Rajneesh",
    "age": 21,
    "course": "Data Analytics"
}

print(student["name"])
