**6. Explain the concept of generators in Python and how they are defined.**

A **generator** is a special type of function that produces values one at a time using the **yield** keyword instead of returning all values at once.


```
def count_up(n):
    for i in range(n):
        yield i

for num in count_up(5):
    print(num)
```
**Generators** save memory because they generate values only when needed.

**9. Explain the purpose and usage of the `map()` function in Python.**

The **map() function** applies a given function to each item of an iterable.


```
# Syntax
map(function, iterable)
# Example

numbers = [1,2,3,4]
square = list(map(lambda x: x*x, numbers))
print(square)
```
